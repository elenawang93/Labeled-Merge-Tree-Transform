### Build Sample Graphs in NetworkX

```python
# sample helper function to integrate output of graph distaince
def _trapezoidIntegration(x, y):
    size = len(x)
    assert size == len(y), "Sizes do not match"
    
    mySum = 0
    
    for i in range(0, size-1):
        mySum += (y[i] + y[i+1]) * (x[i+1] - x[i]) / 2
    
    return mySum
```

```python
# demo graphs
G1 = nx.Graph()
G2 = nx.Graph()
G3 = nx.Graph()

pos1 = {0: (1.8, -3.6), 1: (0, 0), 2: (3, 4.5), 3: (1, 1.5), 4:(1.5, 4)}
pos2 = {0: (1.96, 0.5), 1: (3.96, 1.5)}
pos3 = {0: (2.04, -0.5), 1: (4.04, 0)}

G1.add_nodes_from(pos1.keys())
G2.add_nodes_from(pos2.keys())
G3.add_nodes_from(pos3.keys())

for n1, p1 in pos1.items():
    G1.nodes[n1]['pos'] = p1
    
for n2, p2 in pos2.items():
    G2.nodes[n2]['pos'] = p2
    
for n3, p3 in pos3.items():
    G3.nodes[n3]['pos'] = p3

G1.add_edges_from([(0, 1), (1, 3), (2, 3), (3, 4)])
G2.add_edges_from([(0, 1)])
G3.add_edges_from([(0, 1)])

nx.draw(G1, nx.get_node_attributes(G1, 'pos'), node_color='#00DF3B', edge_color='#00DF3B', with_labels=True)
nx.draw(G2, nx.get_node_attributes(G2, 'pos'), node_color='#0000FF', edge_color='#0000FF', with_labels=True)
nx.draw(G3, nx.get_node_attributes(G3, 'pos'), node_color='#FF0000', edge_color='#FF0000', with_labels=True)
plt.show()
```

### 2 Graph Distaince and Integration

```python
# return values are for more processing
n = 1000
X, Y, _, _ = get2GraphDistance(G1, G2, plot=True, show=False, n=n)
```

```python
dist = _trapezoidIntegration(X, Y)
print(f'dist: {dist}')
```
